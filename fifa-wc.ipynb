# Transformations

```python
def get_condition(con):
    if con.find('extra time') != -1:
        return "ET"
    elif con.find('penalties') != -1:
        return "PEN"
    elif con.find('Golden Goal') != 1:
        return "GG"
    else:
        return ''
    
def get_winner(data):
    h_team = data[0]
    h_goals = data[2]
    a_team = data[1]
    a_goals = data[3]
    cond = data[4]
    stage = data[5]
    if cond != ' ':
        if cond.startswith(' win on penalties'):
            h_goals, a_goals = map(int, cond[cond.find('(') + 1 : cond.find(')')].split(' - '))
        elif cond.find('Golden Goal') == -1:
            return cond.split(' win')[0]
    if h_goals > a_goals:
        return h_team
    elif a_goals > h_goals:
        return a_team
    else:
        return "Draw"
    
def half_time_result(data):
    h_team = data[0]
    h_goals = data[2]
    a_team = data[1]
    a_goals = data[3]
    if h_goals > a_goals:
        return h_team
    elif a_goals > h_goals:
        return a_team
    else:
        return "Draw"
    
def clean_countries(name):
    name = re.sub(r'rn">', '', name)
    name = name.replace("IR Iran", "Iran")
    name = name.replace("Germany FR", "Germany")
    #name = name.replace("German DR", "Germany")
    return name

def clean_initials(name):
    #name = name.replace("GDR", "GER")
    name = name.replace("FRG", "GER")
    return name

def after_half_time(data):
    winner = data[0]
    ht_winner = data[1]
    if winner == ht_winner and winner != "Draw":
        return 'Maintained the lead'
    elif ht_winner != winner:
        return "Bottled the lead"
    else:
        return "Draw till the end"
```

```python
df_match['Win condition'] = df_match[df_match['Win conditions'] != " "]['Win conditions'].apply(lambda x: get_condition(x))
df_match['Win condition'].fillna('', inplace=True)
```

```python
df_match['Home Team Name'] = df_match['Home Team Name'].apply(clean_countries)
df_match['Away Team Name'] = df_match['Away Team Name'].apply(clean_countries)
df_match['Home Team Initials'] = df_match['Home Team Initials'].apply(clean_initials)
df_match['Away Team Initials'] = df_match['Away Team Initials'].apply(clean_initials)
df_match['Match Winner'] = df_match[['Home Team Name','Away Team Name', 'Home Team Goals', 'Away Team Goals', 'Win conditions', 'Stage']].apply(get_winner, axis=1)
df_match['Half time result'] = df_match[['Home Team Name','Away Team Name', 'Half-time Home Goals', 'Half-time Away Goals']].apply(half_time_result, axis=1)
df_match['After Half time'] = df_match[['Match Winner','Half time result']].apply(after_half_time, axis=1)
```

# Creating Countrywise Stats

```python
df_match['Stage'].unique()
```

```python
all_years = df_match['Year'].unique()
all_countries = df_match['Away Team Name'].sort_values().unique()
all_initials = df_match['Away Team Initials'].sort_values().unique().tolist()
df_players['Team Initials'].replace(['GDR', 'FRG'], 'GER', inplace=True)
```

```python
def get_country_results(country):
    df_ctr = df_match[(df_match['Away Team Name'] == country) | (df_match['Home Team Name'] == country)].copy()
    last_row_per_year = df_ctr.groupby(pd.DatetimeIndex(df_ctr.Datetime).year).last()
    ctr_df = last_row_per_year.iloc[:, [0,2,5,8, -4]].copy()

    def get_yearly_result(data):
        yr, stage, ht, at, winner = data['Year'], data['Stage'], data['Home Team Name'], data['Away Team Name'], data['Match Winner']
        opp = at if ht == country else ht
        result, remarks = None, ''
        if stage == "Final":
            result = 'Champions' if winner == country else "Runners-up"
            remarks = ("Champions by beating " + opp + " in the final") if winner == country else ("Lost the final to " + opp)
        elif stage == "Match for third place":
            result = 'Third place' if winner == country else "Fourth place"
            remarks = ("Beaten " + opp + " to finish 3rd") if winner == country else ("Finished 4th behind " + opp)
        elif stage == 'Semi-finals':
            result = 'Fourth place'
            remarks = "Reached the Semifinal"
        elif stage.find('Group') != -1:
            result = 'Group stage'
            remarks = ("Eliminated in the group stages")
        elif stage == 'Round of 16' or stage == 'First round':
            result = 'Round of 16'
            remarks = ("Reached the round of 16") if winner == country else ("Eliminated by " + opp)
        else:
            result = stage
            remarks = ("Reached the " + stage) if winner == country else ("Eliminated by " + opp)
        
        return pd.Series([result, opp, remarks])

    ctr_df['Country'] = country
    initials = df_match[df_match['Away Team Name'] == country]['Away Team Initials'].unique()[0]
    ctr_df['Initials'] = initials
    ctr_df['Attendance'] = df_ctr.groupby('Year')['Attendance'].sum()
    ctr_df[['Result', 'Last opponent', 'Remarks']] = ctr_df.apply(get_yearly_result, axis=1)
    for year in all_years:
        if year not in ctr_df['Year'].values:
            nq_df = pd.DataFrame([[country, initials, year, 'Not qualified', '', 'Did not qualify for the tournament', np.nan]], 
                                 columns=['Country', 'Initials', 'Year', 'Result', 'Last opponent', 'Remarks', 'Attendance'])
            ctr_df = pd.concat([ctr_df, nq_df])
    ctr_df.sort_values('Year', inplace=True)
    ctr_df['Host'] = df_cups['Country'].values
    ctr_df = ctr_df.reset_index(drop=True)
    ctr_df = ctr_df[['Country', 'Initials', 'Host', 'Year', 'Attendance', 'Result', 'Last opponent', 'Remarks']]

    return ctr_df
```

```python
df_country = pd.DataFrame(columns=['Country', 'Initials', 'Host', 'Year', 'Attendance', 'Result', 'Last opponent', 'Remarks'])

for country in all_countries:
    temp = get_country_results(country)
    df_country = pd.concat([df_country, temp])

df_country = df_country.reset_index(drop=True)
```

```python
def get_statistics(dataset):
    dataset['Goals_scored'] = np.nan
    dataset['Goals_conceded'] = np.nan
    dataset['Matches'] = np.nan
    dataset['Wins'] = np.nan
    dataset['Draws'] = np.nan
    dataset['Losses'] = np.nan
    dataset['Win percentage'] = np.nan

    for i, val in dataset.iterrows():
        yr = val['Year']
        ctry = val['Country']
        if (dataset.loc[(dataset['Country'] == ctry) & (dataset['Year'] == yr), 'Result'].values[0] == 'Not qualified'):
            continue
        temp_dt = df_match[((df_match['Home Team Name'] == ctry) | (df_match['Away Team Name'] == ctry)) & (df_match['Year'] == yr)]
        res = temp_dt['Match Winner'].value_counts().to_dict()
        matches = temp_dt.shape[0]
        wins = res[ctry] if ctry in res.keys() else 0
        draws = res['Draw'] if 'Draw' in res.keys() else 0
        losses = matches - wins - draws
        temp_ht_df = df_match[(df_match['Home Team Name'] == ctry) & (df_match['Year'] == yr)]
        temp_at_df = df_match[(df_match['Away Team Name'] == ctry) & (df_match['Year'] == yr)]
        gs_year = temp_ht_df['Home Team Goals'].sum() + temp_at_df['Away Team Goals'].sum()
        gc_year = temp_ht_df['Away Team Goals'].sum() + temp_at_df['Home Team Goals'].sum()

        dataset.iloc[i, -7] = gs_year
        dataset.iloc[i, -6] = gc_year
        dataset.iloc[i, -5] = matches
        dataset.iloc[i, -4] = wins
        dataset.iloc[i, -3] = draws
        dataset.iloc[i, -2] = losses
        dataset.iloc[i, -1] = round(wins / matches * 100, 2)

    dataset['Goals_scored'] = dataset['Goals_scored'].astype('Int8')
    dataset['Goals_conceded'] = dataset['Goals_conceded'].astype('Int8')
    dataset['Matches'] = dataset['Matches'].astype('Int8')
    dataset['Wins'] = dataset['Wins'].astype('Int8')
    dataset['Draws'] = dataset['Draws'].astype('Int8')
    dataset['Losses'] = dataset['Losses'].astype('Int8')
    dataset['Attendance'] = dataset['Attendance'].astype('Int32')
    
    return dataset

df_country = get_statistics(df_country)
```

```python
def get_coaches(data):
    yy, init = data['Year'], data['Initials']
    cur_ids = df_match[(df_match['Year'] == yy) & ((df_match['Home Team Initials'] == init) | (df_match['Away Team Initials'] == init))]['MatchID'].tolist()
    sliced = df_players[(df_players['MatchID'].isin(cur_ids)) & (df_players['Team Initials'] == init)]
    if sliced.shape[0] > 0:
        cur_mgr = sliced.groupby('Coach Name')['MatchID'].nunique().sort_values(ascending=False).index[0]
        return cur_mgr
    return ''
df_country['Coach'] = df_country.apply(get_coaches, axis=1)
```

```python
cols = ['Bottled the lead', 'Maintained the lead', 'Draw till the end']
df_country[cols] = np.nan
```

```python
for country in all_countries:
    for year in all_years:
        temp_ctry = df_match[(df_match['Away Team Name'] == country) | (df_match['Home Team Name'] == country)]
        after_ht = temp_ctry.groupby('Year')['After Half time'].value_counts()
        for i in after_ht.index:
            yr, event, count = i[0], i[1], after_ht[i[0]][0]
            df_country.loc[(df_country['Country'] == country) & (df_country['Year'] == yr), event] = count
            #print(i[0], i[1], after_ht[i[0]][0])

df_country.loc[~df_country['Matches'].isna(), cols] = df_country.loc[~df_country['Matches'].isna(), cols].apply(lambda col: col.fillna(0))
df_country.loc[:, cols] = df_country.loc[:, cols].astype('Int8')
```
