```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.columns
```

```python
le = LabelEncoder()
```

```python
new_df = pd.get_dummies(df.loc[:, ['Region', 'Item Type']], columns=['Region', 'Item Type'])
```

```python
new_df['Sales Channel'] = le.fit_transform(df['Sales Channel'])
```

```python
df['Order Date'][0].year
```

```python
new_df['Order Year'] = df['Order Date'].apply(lambda x: x.year)
new_df['Ship Year'] = df['Ship Date'].apply(lambda x: x.year)
```

```python
new_df[['Units Sold', 'Unit Price','Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']] = df[['Units Sold', 'Unit Price','Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']]
```
